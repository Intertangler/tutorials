#Guarantee 9 out of 10 prisoners' survival using the cyclic group $\boldsymbol{C}_3$

Group theory is the branch of mathematics that deals with deterministic, reversible operations on $groups$ of discrete objects. Many other mathematical objects such as graphs, sudoku puzzles, rubiks cubes, and algebraic expressions can often be described in terms of group theory.

The 10 prisoners 3 colors puzzle can be solved using a very simple algorithm that would not require each prisoner to have a pen and paper, calculators, computer programming, or even knowledge of division implement the solution this way. The key is to use the cyclic group $\boldsymbol{C}_3$ shown as a Caley diagram on the left side of the image below, simply 3 connected actions or operations associated with each color of hat which map to each other the way a clock would. ie a series of these actions is reducible to one of the actions in the same way that 8:00 am plus 20 hours is just 4:00 am. The clock is a cyclic group.

In our case, the last prisoner in the line who sees all 9 prisoners ahead begins in an initial state (state 0) agreed upon by everyone before the lineup. The lineup ahead is then treated as a series of operations, with each color corresponding to the action: operation 0 - yellow: move 0 states ahead, operation 1 - blue: move 1 state ahead, and operation 2 - red: move 2 states ahead. The moves/operations and their results can be thought about visually via the diagram on the right. 

Most importantly, a series of operations will deterministically put you in a particular final state. For instance: operation 2 followed by operation 1 leads you to state 0. Operation 2 followed by operation 2 again leads you to state 1. This is analogous to series addition of time intervals on the 12 hour clock, eg from 2:00, 4 hr + 8 hr + 2 hr = 4:00

The 3 operations are mapped conveniently such that if you made one operation in a series unknown, but were given the final state, you could deduce the operation knowing the expected final state without the mystery operation. No operations are redundant in other words.

So in practice this looks like the following:
Prisoner 10 sees the following sequence: ybybryryb or the operations: 010120201. Beginning in state 0, the prisoner works through the operations ending finally with state 1 - he announces blue to indicate this (for simplicity I've mapped the states and operation numbers to the same colors). Hearing 1 behind him, the 9th prisoner knows that the sequence in front of him 10120201 plus the operation of his own hat must put us in state 1. The 9th prisoner performs the series of operations on the cyclic group which reduces to the final state 1. The only binary operation with 1 that yields another 1 is operation 0, so prisoner 9 deduces his hat to be yellow. The pattern continues down the line with each prisoner knowing the final state after performing the series of known operations plus the certain knowledge that his own hat represents a unique operation needed to reach the state announced by the prisoner at the back of the line. 



#binary operations and unique results

A couple principles are at work here. The cyclic 3 state group enables reduction of a series of operations - essential for our situation where limited information can be conveyed. However, cyclicness is not enough. Each prisoner is left to work out the inverse to a binary operation, like solving the algebraic expression 1 + x = 2. One way to visualize binary operations on groups is with a table. On one axis we have one operation, and on the other we have a second operation. Filling the table are the resultant states reached by combining the two operations:


\begin{equation}
 \begin{matrix}
  & _{0} & _{1} & _{2} \\
  _{0}& 0 & 1 & 2 \\
  _{1}& 1 & 2 & 0 \\
  _{2}& 2 & 0 & 1
 \end{matrix}
\end{equation}

Notice how, like a completed sudoku puzzle, no row contains an element whose position is repeated in another row, and likewise for the columns. Here we see though that it is possible to rearrange the results in other uniquely-mapped ways that still provide the deductive certainity we need. eg:

\begin{equation}
 \begin{matrix}
  & _{0} & _{1} & _{2} \\
  _{0}& 1 & 2 & 0 \\
  _{1}& 2 & 0 & 1 \\
  _{2}& 0 & 1 & 2
 \end{matrix}
\end{equation}

This is similar to remapping the numbers to different colors. 

#isomorphism with base 3 counting 

Other routes to solving this puzzle are fundamentally connected. Say we decided to count in base 3 through every one of the 3^9 distinguishable permutations potentially seeable by the last prisoner. Assign a unique label to each triplet that differs by only 1 digit. 

000000000 = "yellow"

000000001 = "blue"

000000002 = "red"

000000010 = "yellow"

000000011 = "blue"

000000012 = "red"

000000020 = "yellow"

...and so on.

If every prisoner could just remember these, we'd be set. But another way is to use modulo 3 in base 3 (eg 010120120 % 10 = 0). Alternatively, convert to base 10 and modulo by 3. For example 12 % 3 = 0, 13 % 3 = 1, 14 % 3 = 2 15 % 3 = 0. Notice how modulo achieves the same reduction of series couting operations as series cyclic group operations. Herein lies the connection. For example base 10 counting on our fingers is essentially the cyclic group $C_{10}$. If I ask you to count to 15 on your fingers, you can do it, not because you can sprout additional fingers but because you can count cyclically.
